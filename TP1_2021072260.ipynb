## PARTE 1 - Transformar Algebra Relacional para SQL
Abaixo, seguem as consultas que voce terá de transformar de Algebra Relacional para SQL. Altere **APENAS** o espaço entre as aspas. Dentro desse espaço é que você deverá digitar o comando de execução da consulta.

### 1 - $\pi_{nome\_fornecedor} \: \sigma_{uf  =  'SP'}$ (favorecido)




```python
a1 = """
SELECT 
    nome_fornecedor 
FROM 
    favorecido 
WHERE  
    uf = 'SP'
"""
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(a1, "a1")
```

### 2 - $\pi_{codigo\_empenho, valor\_empenho} \: \sigma_{valor\_empenho  >  1000000}$ (empenho)

```python
a2 = """
SELECT 
  codigo_empenho, valor_empenho
FROM 
  empenho
WHERE 
  valor_empenho > 1000000;
"""
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(a2, "a2")
```

### 3 - $\pi_{nome\_fornecedor, valor\_empenho} \: \sigma_{valor\_empenho  <  1000}$ (favorecido $\bowtie$ empenho )

```python
a3 = '''
SELECT favorecido.nome_fornecedor, empenho.valor_empenho
FROM favorecido
INNER JOIN empenho ON favorecido.id_favorecido = empenho.id_favorecido
WHERE empenho.valor_empenho < 1000
'''
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(a3, "a3")
```

### 4 - $\pi_{nome\_unidade, valor\_empenho} \: \sigma_{codigo\_unidade = 152005}$ (empenho $\bowtie$ unidade)


```python
a4 = '''
SELECT unidade.nome_unidade, empenho.valor_empenho
FROM empenho
INNER JOIN unidade ON unidade.id_unidade = empenho.id_favorecido
WHERE unidade.codigo_unidade = 152005 
'''
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(a4, "a4")
```

### 5 - $\pi_{id\_unidade, nome\_unidade} \: \sigma_{nome\_orgao = Ministério\  da\ Economia}$ (orgao $\bowtie$ unidade)


```python
a5 = """
SELECT unidade.id_unidade, unidade.nome_unidade
FROM orgao
INNER JOIN unidade ON unidade.id_unidade = orgao.id_orgao
WHERE orgao.nome_orgao = 'Ministério da Economia'
"""
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(a5, "a5")
```

## PARTE 2 - Consultas SQL
Abaixo, seguem as consultas que voce terá de executar. Altere **APENAS** o espaço entre as aspas. Dentro desse espaço é que você deverá digitar o comando de execução da consulta.

### 1 - Liste os campos id_empenho, observacao e data_emissao para todos os empenhos com data de emissão entre 2020-01-01 e 2020-12-31:

```python
q1 = '''
SELECT id_empenho, observacao, data_emissao
FROM empenho
WHERE data_emissao BETWEEN '2020-01-01' AND '2020-12-31'
'''
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(q1, "q1")
```

### 2 - Liste as 3 unidades federativas (uf) com menos favorecidos cadastrados e a quantidade de favorecidos destes Estados:

```python
q2 = '''
SELECT uf, COUNT(id_favorecido) as total_favorecidos
FROM favorecido
GROUP BY uf
ORDER BY total_favorecidos ASC
LIMIT 3;
'''
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(q2, "q2")
```

### 3 - Liste o nome das unidades que começam com a palavra SUPERINTENDENCIA e o nome do orgão ao qual pertencem:

```python
q3 = '''
SELECT orgao.nome_orgao, unidade.nome_unidade
FROM unidade
INNER JOIN orgao ON unidade.id_orgao = orgao.id_orgao
WHERE unidade.nome_unidade LIKE 'SUPERINTENDENCIA%'
'''
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(q3, "q3")
```

### 4 - Liste os códigos de empenho que tenham a palavra COBRIR no campo observacao:
**OBS** (pesquisar a palavra em caixa alta).

```python
q4 = '''
SELECT codigo_empenho
FROM empenho
WHERE observacao LIKE '%COBRIR%'
'''
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(q4, "q4")
```

### 5 - Liste, em ordem crescente, os nomes de todos os fornecedores que tenham mais de 1 empenho, exibindo também a quantidade de empenhos relacionados à cada um deles:

```python
q5 = '''
SELECT favorecido.nome_fornecedor, COUNT(DISTINCT empenho.id_empenho) AS quantidade_empenhos
FROM empenho
INNER JOIN favorecido ON empenho.id_favorecido = favorecido.id_favorecido
GROUP BY favorecido.nome_fornecedor
HAVING COUNT(DISTINCT empenho.id_empenho) > 1
ORDER BY favorecido.nome_fornecedor ASC
'''
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(q5, "q5")
```

### 6 - Liste, em ordem crescente, o nome das unidades e a soma do valor de todos os itens pertencentes a cada uma delas:

**Desafio**: Para tratar o erro de soma dos flutuantes na consulta, você pode reduzir o numero de casas decimais que serão exibidos.

```python
q6 = '''
SELECT unidade.nome_unidade, SUM(item.valor_total_item) AS soma_valores
FROM unidade 
INNER JOIN empenho ON unidade.id_unidade = empenho.id_unidade
INNER JOIN item_empenho AS item ON empenho.id_empenho  =  item.id_empenho
GROUP BY unidade.nome_unidade
ORDER BY soma_valores ASC
'''
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(q6, "q6")
```

### 7 - Liste o nome dos orgãos que tenham mais de 10 favorecidos e a quantidade de favorecidos relacionados à cada orgão:

```python
q7 = '''
SELECT orgao.nome_orgao, COUNT(favorecido.id_favorecido) AS qtd_favorecidos
FROM orgao
INNER JOIN unidade ON orgao.id_orgao = unidade.id_orgao
INNER JOIN empenho ON unidade.id_unidade = empenho.id_unidade
INNER JOIN favorecido ON empenho.id_favorecido = favorecido.id_favorecido
GROUP BY orgao.nome_orgao
HAVING COUNT(favorecido.id_favorecido) > 10
'''
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(q7, "q7")
```

### 8 - Liste o id de todos os empenhos que tenham alguma relação com o orgão Ministério da Saúde:

```python
q8 = '''
SELECT empenho.id_empenho
FROM empenho
INNER JOIN unidade ON empenho.id_unidade = unidade.id_unidade
INNER JOIN orgao ON unidade.id_orgao = orgao.id_orgao
WHERE orgao.nome_orgao = 'Ministério da Saúde'
'''
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(q8, "q8")
```

### 9 - Liste o nome das unidades que possuem favorecidos em MG:

```python
q9 = '''
SELECT DISTINCT unidade.nome_unidade
FROM unidade
INNER JOIN empenho ON unidade.id_unidade = empenho.id_unidade
INNER JOIN favorecido ON empenho.id_favorecido = favorecido.id_favorecido
WHERE favorecido.uf = 'MG'
'''
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(q9, "q9")
```

### 10 - Liste, em ordem crescente, a descriçãoo dos itens de empenho relacionados ao orgão do Ministério da Saúde:

```python
q10 = '''
SELECT item_empenho.descricao
FROM item_empenho
JOIN empenho ON empenho.id_empenho = item_empenho.id_empenho
JOIN unidade ON unidade.id_unidade = empenho.id_unidade
JOIN orgao ON orgao.id_orgao = unidade.id_orgao
WHERE orgao.nome_orgao = 'Ministério da Saúde'
ORDER BY item_empenho.descricao ASC
'''
#NÃO ALTERE AS LINHAS ABAIXO
runQuery(q10, "q10")
```
